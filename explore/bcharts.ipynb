```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-whitegrid')

plt.rc('figure', autolayout=True)
plt.rc('axes', labelweight='bold', labelsize='large',
        titleweight='bold', titlesize=18, titlepad=10)
```

```python
import pandas as pd
import scipy.stats as st

# Imports
source = lambda x : f'../../star_dataset/{x}'

X_full = pd.read_csv(source('Star39552_balanced.csv'))
X_full.head()
```

```python
target = 'TargetClass'
x_var = 'Amag'
y_var = 'B-V'

def create_plot(x_var, y_var):
    max_x = X_full[x_var].max()
    max_y = X_full[y_var].max()
    x_giant = X_full[X_full[target] == 1][x_var] # / max_x
    y_giant = X_full[X_full[target] == 1][y_var] # / max_y
    x_dwarf = X_full[X_full[target] == 0][x_var] # / max_x
    y_dwarf = X_full[X_full[target] == 0][y_var] # / max_y
    plt.scatter(x_giant, y_giant, s=0.1, color='red', label="Giant (red)")
    plt.scatter(x_dwarf, y_dwarf, s=0.1, color='green', label="Dwarf (green)")
    plt.legend()
    plt.title("Dwarfs and Giants")
    plt.ylabel(y_var)
    plt.xlabel(x_var)

create_plot(x_var, y_var)
```

```python
x_var = 'Vmag'
y_var = 'B-V'
create_plot(x_var, y_var)
```

```python
x_var = 'Vmag'
y_var = 'Amag'
create_plot(x_var, y_var)
```
